```python
import pandas as pd
from matplotlib import pyplot as plt 
import datetime as dt
import seaborn as sns
```

```python
df = pd.read_csv("Amazon Sales data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
df.head()
```

```python
sns.lineplot(data = df, x = df['Order Year'], y = df['Units Sold'], color = 'green')
plt.title("Sales Performance Analysis: Trends in Units Sold (2010-2017)")
plt.grid()
```

```python
sns.lineplot(data = df, x = df['Order Year'], y = df['Total Revenue'], color='green')
plt.title('Revenue Trends Over 2010-2017: An Analysis')
plt.grid()
```

```python
data_preproc = pd.DataFrame({
    'Year': df['Order Year'], 
    'Total Revenue': df['Total Revenue'],
    'Total Cost' : df['Total Cost']
    })
```

```python
sns.lineplot(x='Year', y='value' , hue='variable', data=pd.melt(data_preproc, ['Year']), errorbar=None)
plt.legend()
plt.grid()
plt.title("Total Revenue vs Total Cost 2010-2017")
plt.show()
```

```python
df.head()
```

```python
sns.boxplot(data = df, x = df['Order Month'], y = df['Total Profit'])
plt.grid()
plt.title("Monthly Total Profit: A Box Plot Analysis (2010-2017)")
```

```python
df_2010 = df[df['Order Year']==2010]
```

```python
df_2010.head()
```

```python
df_2010 = df_2010.sort_values(by='Total Profit', ascending = False)
```

```python
df_2010.head()
```

```python
sns.barplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Profit'])
plt.grid()
plt.title("Monthly Profit Trends in 2010: A Bar Plot Analysis")
plt.show()
```

```python
sns.lineplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Revenue'], color = 'green')
plt.grid()
plt.title("Monthly Revenue Trends in 2010: A Line Plot Analysis")
```
